```python
import pandas as pd
import random
```

```python
# Define the range for each attribute
num_entities = 4000

wallet_numbers = ['1' + ''.join([str(random.randint(0, 9)) for _ in range(9)]) for _ in range(num_entities)]


address_percentages = {
    "Urban": 0.8,
    "Rural": 0.2
}

# Create a list of addresses based on their percentages
addresses = random.choices(list(address_percentages.keys()), k=num_entities, weights=list(address_percentages.values()))


genders = random.choices(["Male", "Female"], k=num_entities)

balances = [round(random.uniform(2000,10000), 2) for _ in range(num_entities)]

balance_deviations = [round(random.uniform(50, 80), 2) for _ in range(num_entities)]
```

```python
# Define age ranges for different occupations
age_ranges = {

"Business": (25, 50),
    "Private Service": (20,60),
    "Govt. Service": (20,60),
    "House Wife": (30, 60),
    "Student": (18, 25)


}
```

```python
# Define the possible occupations and their respective percentages
occupations = [
     ("Business", 0.25),
    ("House Wife", 0.15),
    ("Student", 0.5),
    ("Private Service",0.05),
    ("Govt. Service",0.05)


]
```

```python
# Create lists for occupations and ages based on the defined age ranges
occupation_list = []
age_list = []

for occupation, percentage in occupations:
    occupation_count = int(num_entities * percentage)
    occupation_list.extend([occupation] * occupation_count)
    age_min, age_max = age_ranges[occupation]
    age_list.extend([random.randint(age_min, age_max) for _ in range(occupation_count)])

# Shuffle the occupation and age lists to randomize the assignment
combined = list(zip(occupation_list, age_list))
random.shuffle(combined)
occupation_list, age_list = zip(*combined)
```

```python
# Create a DataFrame
data = {
    "Wallet No": wallet_numbers,
    "Address": addresses,
    "Balance": balances,
    "Age": age_list,
    "Occupation": occupation_list,
    "Loan Approval": "Eligible",
    "Balance Deviation (CV)": balance_deviations,

}
```

```python
df = pd.DataFrame(data)
```
